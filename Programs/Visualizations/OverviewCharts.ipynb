# Overview Chart Creation
Sample of clustering results and Number of aspects / clusters per category

```python
import pandas as pd

# pyspark
from pyspark import SparkConf, SparkContext
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```python
# init spark
spark = SparkSession \
    .builder \
    .appName("OverviewViz") \
    .getOrCreate()

base_path = "gs://core_bucket_abell/aspect_ranking"
```

```python
categories = ['All Electronics', 'Electronics','Home & Kitchen', 'Home Improvement', 
 'Industrial & Scientific', 'Health & Personal Care',  'Camera & Photo', 'Software',
 'Cell Phones & Accessories','Arts', 'Computers', 'Office & School Supplies',
 'GPS & Navigation','Car Electronics', 'Toys & Games', 'Kitchen & Dining','Tools & Home Improvement',
 'Baby', 'All Beauty', 'Baby Products', 'Office Products','Appliances',
'Amazon Fashion','Patio', 'Sports & Outdoors', 'Video Games',
 'Clothing', 'Musical Instruments', 'Automotive', 'Luxury Beauty']
```

### First Get Cluster Results sample

```python
sample_df = spark.read.parquet(base_path + "/sentiment_analysis/Sentiments All Electronics.parquet")\
    .select("split_aspect", "key_aspect").limit(10000).toPandas()
sample_df.head()
```

```python
sample_df.iloc[67:80,:]
```

### Now number of unique aspects and clusters per category

```python
row = []
for cat in categories:
    df = spark.read.parquet(base_path + "/sentiment_analysis/Sentiments " + cat + ".parquet")
    num_aspects = df.select("split_aspect").distinct().count()
    num_clusters = df.select("key_aspect").distinct().count()
    
    print(cat, str(num_aspects), str(num_clusters))
    row.append([cat, num_aspects, num_clusters])
```

```python
pd.DataFrame(row, columns = ["Category", "No. of Unique Aspects", "No. of Clusters"]).to_csv(base_path + "/Cat Overview.csv", index = False)
```
